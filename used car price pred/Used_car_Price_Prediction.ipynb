```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
X=pd.read_excel('Data_Train.xlsx')
```

```python
X.head(5)
```

```python
print(X.shape)
print(X.info())
```

```python
le=LabelEncoder()
#drop_columns=['New_Price','Name']
X=X.drop('New_Price',axis=1)
```

```python
X.shape
```

```python
X['Name']=le.fit_transform(X['Name'])
X['Location']=le.fit_transform(X['Location'])
X['Fuel_Type']=le.fit_transform(X['Fuel_Type'])
X['Transmission']=le.fit_transform(X['Transmission'])
X['Owner_Type']=le.fit_transform(X['Owner_Type'])
X['Engine']=X['Engine'].apply(lambda x: str(x)[:-3])
X['Mileage']=X['Mileage'].apply(lambda x:str(x)[:-5])
X['Power']=X['Power'].apply(lambda x: str(x)[:-3])
```

```python
X.head(5)
```

```python
a = X['Power']
for i in range(6019):
    print(a[i])
    try:
        if 'null' in a[i]:
            a[i] = np.NAN
    except:
        pass
```

```python
X['Engine'] = pd.to_numeric(X['Engine'])
X['Mileage'] = pd.to_numeric(X['Mileage'])
X['Power'] = pd.to_numeric(X['Power'])
```

```python
X['Mileage']=X.fillna(X['Mileage'].mean())
X['Engine']=X.fillna(X['Engine'].mean())
X['Power']=X.fillna(X['Power'].mean())
X['Seats']=X.fillna(X['Seats'].mean())
```

```python
X.info()
```

```python
Y=X['Price']
X=X.drop('Price',axis=1)
```

```python
Y.shape
```

```python
print(X.info())
print(Y.head())
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
ln = LinearRegression(normalize=True)
```

```python
x=(X-X.mean())/X.std()
y=(Y-Y.mean())/Y.std()
```

```python
from sklearn.model_selection import train_test_split
xt,xv,yt,yv=train_test_split(x,y,test_size=0.2)
```

```python
ln.fit(xt,yt)
```

```python
ln.intercept_
```

```python
ln.coef_
```

```python
train_acc=ln.score(xt,yt)
val_acc=ln.score(xv,yv)
```

```python
print('training accuracy is {0}\n validation accuracy is {1}'.format(train_acc,val_acc))
```

## Decision tree

```python
from sklearn import tree
```

```python
m=tree.DecisionTreeRegressor(max_depth=12)
```

```python
m.fit(xt,yt)
```

```python
train_a=m.score(xt,yt)
val_a=m.score(xv,yv)
```

```python
print('training accuracy is {0}\n validation accuracy is {1}'.format(train_a,val_a))
```

## Neural Network

```python
from keras.layers import *
from keras.models import Sequential
from keras.optimizers import Adam
from keras.utils import np_utils
```

```python
#from sklearn.model_selection import train_test_split
```

```python
#xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
```

```python
model=Sequential()
model.add(Dense(128,activation='tanh',input_shape=(11,)))
model.add(Dense(64,activation='tanh'))
model.add(Dense(10,activation='tanh'))
model.add(Dense(1,activation='linear'))
model.summary()
```

```python
#adam=Adam(decay=0.00003)

model.compile(loss='mae',optimizer='adam')
```

```python
hist=model.fit(x,y,epochs=50,shuffle=False,validation_split=0.2)
```

```python
plt.plot(hist.history['loss'],label='training_loss')
plt.plot(hist.history['val_loss'],label='validation_loss')
plt.legend()
plt.show()
```

```python
model.evaluate(x,y)
```
